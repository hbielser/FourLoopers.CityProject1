```python
#dependencies
import pandas as pd

import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as sem
from scipy.stats import linregress
import numpy as np
from sklearn import datasets
```

```python
# import cleaned study data from csv
county_health_df = "Resources/FullData.csv"
county_health_df = pd.read_csv(county_health_df)
county_health_df.head()
```

```python
#Create variables to graph
prev_hosp = county_health_df["Preventable Hospitalization Rate"]
food_index = county_health_df["Food Environment Index"]
years_ll = county_health_df["Years of Potential Life Lost Rate"]
obesity = county_health_df["% Adults with Obesity"]
unempl = county_health_df["% Unemployed"]
income_ratio = county_health_df["Income Ratio"]
fair_health = county_health_df['% Fair or Poor Health']
ph_unhealthy = county_health_df['Average Number of Physically Unhealthy Days']
mh_unhealthy = county_health_df['Average Number of Mentally Unhealthy Days']
pcp_ratio = county_health_df['Primary Care Physicians Ratio']
pop = county_health_df['Population']
social = county_health_df['Social Association Rate']
smokers = county_health_df['% Smokers']
drinking = county_health_df['% Excessive Drinking']
grad = county_health_df['High School Graduation Rate']
```

```python
# assess relationship of variables between two outcomes: preventable hosp and poor health
x_values = prev_hosp
y_values = fair_health

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,5),fontsize=15,color="red")

print(f"The r-squared is: {rvalue**2}")
#add labels
plt.xlabel('prev_hosp')
plt.ylabel('fair_health')
plt.title(f"prev_hosp and fair_health")
plt.savefig("Prev Hosp & Fair Health.png")
plt.show()
```

```python
# pick top and bottom counties of fair or poor health outcomes and create study dataframe

n_cities=20

top_n = county_health_df.nlargest(n_cities, ['% Fair or Poor Health'])
bottom_n = county_health_df.nsmallest(n_cities, ['% Fair or Poor Health'])
study_df = pd.concat([top_n,bottom_n])

study_df = study_df.reset_index(drop=True)
study_df
```

```python
study_df.to_csv("Resources/PoorHealth_StudyData.csv")
```

```python
top_grouped = top_n.groupby(['State']).count()
county_count_top = top_grouped['County']
county_count_top
```

```python
labels = top_grouped.index
numbers = top_grouped['County']
county_count_top.plot(kind='pie', labels= numbers, title="Number of Counties with Highest % of Poor Health Population by State", fontsize=10)
plt.legend(loc = "center left", labels= labels,bbox_to_anchor=(1,0,.5,.6))
plt.axis("equal")
def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)
plt.savefig("Counties with Highest % of Poor Health.png")
plt.show()
```

```python
bottom_grouped = bottom_n.groupby(['State']).count()
county_count_bottom = bottom_grouped['County']
county_count_bottom
```

```python
labels = bottom_grouped.index
numbers = bottom_grouped['County']
county_count_bottom.plot(kind='pie', labels= numbers, title="Number of Counties with Lowest % of Poor Health Population by State", fontsize=10)
plt.legend(loc = "center left", labels= labels,bbox_to_anchor=(1,0,.5,.6))
plt.axis("equal")
def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)
plt.savefig("Counties with Lowest % of Poor Health.png")
plt.show()
```
